```python
#giving band info
#import bands as separate 1 band raster
#making variables for bands needed for the index equation
Band5 = rasterio.open("C:/Users/ASMITA/Documents/Python_rasteranalysis/kawa_space/place1/data/raster/LC08_L2SP_148039_20201005_20201016_02_T1_SR_B5.TIF")
Band4 = rasterio.open("C:/Users/ASMITA/Documents/Python_rasteranalysis/kawa_space/place1/data/raster/LC08_L2SP_148039_20201005_20201016_02_T1_SR_B4.TIF")
```

#msavi2 formula is
#MSAVI2 = 1/2 * ((2*(NIR+1)) - (((2*NIR)+1)**2 - 8*(NIR-red))**1/2)

```python
#generate nir and red objects as arrays in float64 format
red = Band4.read(1).astype('float64')
nir = Band5.read(1).astype('float64')
```

```python
a = (2*(nir+1))
```

```python
b = ((2*nir)+1)
```

```python
c = (8*(nir-red))
```

```python
d = b**2
```

```python
e = (1/2)
```

```python
f = (d-c)**e
```

```python
g = (a-f)
```

```python
msavi2 = e*g
```

```python
#this is the final msavi2 array result
msavi2
```

```python
##exporting msavi2 image
msaviImage = rasterio.open("C:/Users/ASMITA/Documents/Python_rasteranalysis/kawa_space/place1/data/msaviPlace1.tif",'w',driver='Gtiff',width=Band4.width, height=Band4.height, count=1, crs=Band4.crs, transform=Band4.transform,dtype='float64')
msaviImage.write(msavi2,1)
msaviImage.close()
```
